```python
df = pd.read_csv('data/cleaned_Shoping_behavior_dataset.csv')
df.head(3)
```

```python
# Creating age groups
age_bins = [0, 18, 25, 35, 45, 55, 65, 100]
age_labels = ['<18', '18-25', '26-35', '36-45', '46-55', '56-65', '65+']
df["Age Group"] = pd.cut(df["Age"], bins=age_bins, labels=age_labels, right=False)

#purchases by age group
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="Age Group", palette="Set1", order=age_labels)
plt.title("Purchase Count by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of Purchases")
plt.show()

plt.savefig("visuals/PurhaseByAgeGroup.png", dpi=300, bbox_inches='tight')
```

```python
#gender distribution of purchases
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="Gender", palette="Set1")
plt.title("Purchases by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Purchases")
plt.show()

plt.savefig("visuals/PurchaseByGender.png", dpi=300, bbox_inches='tight')
```

```python
#frequency of each payment method
payment_counts = df["Payment Method"].value_counts()

plt.figure(figsize=(5, 5))
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Frequently Used Payment Methods")
plt.axis('equal') 
plt.show()

plt.savefig("visuals/PaymentMethods.png", dpi=300, bbox_inches='tight')
```

```python
# Count frequency of each shipping method
shipping_counts = df["Shipping Type"].value_counts()

# Plot pie chart of shipping methods
plt.figure(figsize=(5, 5))
plt.pie(shipping_counts, labels=shipping_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Frequently Used Shipping Methods")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

plt.savefig("visuals/FrequentyUsedShippingMethods.png", dpi=300, bbox_inches='tight')
```

```python
# Count purchases by season
season_counts = df["Season"].value_counts()

#purchases by season
plt.figure(figsize=(6, 4))
sns.barplot(x=season_counts.index, y=season_counts.values, palette="Set1")
plt.title("Purchases by Season")
plt.xlabel("Season")
plt.ylabel("Number of Purchases")
plt.show()

plt.savefig("visuals/PurchaseBySeason.png", dpi=300, bbox_inches='tight')
```

```python
# Group by season and sum the purchase amounts
season_sales = df.groupby("Season")["Purchase Amount (USD)"].sum().sort_values(ascending=False)

# Plot total sales by season
plt.figure(figsize=(6, 4))
sns.barplot(x=season_sales.index, y=season_sales.values, palette="Set1")
plt.title("Total Sales by Season")
plt.xlabel("Season")
plt.ylabel("Totavisuals/l Purchase Amount (USD)")
plt.show()

plt.savefig("visuals/TotalSalesbySeason.png", dpi=300, bbox_inches='tight')
```

```python
# Count purchases by item size
size_counts = df["Size"].value_counts()

# Plot purchases by size
plt.figure(figsize=(6, 4))
sns.barplot(x=size_counts.index, y=size_counts.values, palette="Set1")
plt.title("Purchases by Item Size")
plt.xlabel("Size")
plt.ylabel("Number of Purchases")
plt.show()

plt.savefig("visuals/PurchaseByItems.png", dpi=300, bbox_inches='tight')
```

```python
# Count the most sold items
item_counts = df["Item Purchased"].value_counts().head(10)

# Most sold items
plt.figure(figsize=(6, 4))
sns.barplot(x=item_counts.values, y=item_counts.index, palette="Set1")
plt.title("Top 10 Most Sold Items")
plt.xlabel("Number of Purchases")
plt.ylabel("Item Purchased")
plt.tight_layout()
plt.show()

plt.savefig("visuals/Top10ItemsSold.png", dpi=300, bbox_inches='tight')
```

```python
# Count purchases by category
category_counts = df["Category"].value_counts()

# Plot purchase count by category
plt.figure(figsize=(6, 4))
sns.barplot(x=category_counts.index, y=category_counts.values, palette="Set1")
plt.title("Purchases by Category")
plt.xlabel("Category")
plt.ylabel("Number of Purchases")
plt.tight_layout()
plt.show()

plt.savefig("visuals/Purchase_by_category.png", dpi=300, bbox_inches='tight')
```

```python
# Crosstab of Category vs. Season
category_season_crosstab = pd.crosstab(df["Category"], df["Season"])

# Plotting grouped bar chart
category_season_crosstab.plot(kind='bar', figsize=(6, 4), colormap='tab10')
plt.title("Item Categories Sold by Season")
plt.xlabel("Category")
plt.ylabel("Number of Purchases")
plt.legend(title="Season")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

plt.savefig("visuals/Category_Sold_by_season.png", dpi=300, bbox_inches='tight')
```
